## Data Preprocessing
** **
In this section, I will preprocess my Netflix Viewing history data which I got from Netflix. I will load the data into a data frame and perform the necessary checks and cleans.

```python
#importing the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#loading the data

df = pd.read_csv('ViewingActivity.csv')

display(df)
display(df.shape)
```

```python
# Filter the data where Profile Name is 'Zeynep'
zeynep_data = df[df['Profile Name'] == 'Zeynep']

# Display the filtered data
display(zeynep_data)
```

```python
import pandas as pd

# Drop columns if they exist in the DataFrame
columns_to_drop = ['Attributes', 'Supplemental Video Type']

existing_columns = zeynep_data.columns.tolist()
columns_to_drop = [col for col in columns_to_drop if col in existing_columns]

zeynep_data = zeynep_data.drop(columns=columns_to_drop, axis=1)

# Display the modified DataFrame
display(zeynep_data)
```

```python
# Computing descriptive statistics for the dataset
zeynep_data.describe()
```

```python
# Obtaining information about the dataset
zeynep_data.info()
```

```python
# Get data types of each column
data_types = zeynep_data.dtypes

# Display data types
print(data_types)
```

```python
# Convert 'Start Time' column to datetime with UTC
zeynep_data['Start Time'] = pd.to_datetime(zeynep_data['Start Time'], utc=True)

# Check the data types after conversion
print(zeynep_data.dtypes)
```

```python
# Change 'Start Time' column into the dataframe's index
zeynep_data = zeynep_data.set_index('Start Time')

# Convert from UTC timezone to Eastern Time (assuming 'Europe/Istanbul' is your desired timezone)
zeynep_data.index = zeynep_data.index.tz_convert('Europe/Istanbul')

# Reset the index so that 'Start Time' becomes a column again
zeynep_data = zeynep_data.reset_index()

# Display the first row to check if the operations worked
display(zeynep_data)
```

```python
# Convert 'Duration' column to timedelta
zeynep_data['Duration'] = pd.to_timedelta(zeynep_data['Duration'])
zeynep_data['Bookmark'] = pd.to_timedelta(zeynep_data['Bookmark'])

# Check the data types after conversion
print(zeynep_data.dtypes)
```

## Model Evaluation
** **



In this section, we will use 2 models: linear regression, cluster analysis

```python
corona_avg_viewing_time = corona_data['Duration'].mean()
after_corona_avg_viewing_time = after_corona_data['Duration'].mean()
print(f"Average viewing time during Corona: {corona_avg_viewing_time}")
print(f"Average viewing time after Corona: {after_corona_avg_viewing_time}")
```

```python
corona_total_viewing_time = corona_data['Duration'].sum()
after_corona_total_viewing_time = after_corona_data['Duration'].sum()
print(f"Total viewing time during Corona: {corona_total_viewing_time}")
print(f"Total viewing time after Corona: {after_corona_total_viewing_time}")
```

```python
from scipy.stats import f_oneway

# Extract duration data for corona and after corona periods
corona_durations = corona_data['Duration']
after_corona_durations = after_corona_data['Duration']

# Perform ANOVA
f_statistic, p_value = f_oneway(corona_durations, after_corona_durations)

# Interpret Results
if p_value < 0.05:  # Considering a significance level of 0.05
    print("The difference in duration (viewing times) between periods is statistically significant.")
else:
    print("There is no statistically significant difference in duration (viewing times) between periods.")
```

```python
from scipy import stats
# Perform t-test for average viewing times
corona_duration_minutes = corona_data['Duration'].dt.total_seconds() 
after_corona_duration_minutes = after_corona_data['Duration'].dt.total_seconds()

# Perform t-test for average viewing times
t_stat, p_value = stats.ttest_ind(corona_duration_minutes, after_corona_duration_minutes)

# Display the t-statistic and p-value
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")

# Interpret the results
if p_value < 0.05:  # Considering a significance level of 0.05
    print("The difference in average viewing times is statistically significant.")
else:
    print("There is no statistically significant difference in average viewing times.")
```

```python
#'Duration' columns are in timedelta format
corona_duration_minutes = corona_data['Duration'].dt.total_seconds() / 60
after_corona_duration_minutes = after_corona_data['Duration'].dt.total_seconds() / 60

# Perform t-test for total viewing times
t_stat, p_value = stats.ttest_ind(corona_duration_minutes, after_corona_duration_minutes)

# Display the t-statistic and p-value
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")

# Interpret the results
if p_value < 0.05:  # Considering a significance level of 0.05
    print("The difference in total viewing times is statistically significant.")
else:
    print("There is no statistically significant difference in total viewing times.")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder

# Assuming 'Duration' columns are in timedelta format
corona_data.loc[:, 'Duration_minutes'] = corona_data['Duration'].dt.total_seconds() 
after_corona_data.loc[:, 'Duration_minutes'] = after_corona_data['Duration'].dt.total_seconds() 

# Creating a label for the time period (corona and after corona)
corona_data.loc[:, 'Period'] = 'Corona'
after_corona_data.loc[:, 'Period'] = 'After Corona'

# Combining the data
combined_data = pd.concat([corona_data, after_corona_data])

# Encoding the 'Period' column
label_encoder = LabelEncoder()
combined_data.loc[:, 'Period_encoded'] = label_encoder.fit_transform(combined_data['Period'])

# Performing linear regression
X = combined_data[['Period_encoded']]
y = combined_data['Duration_minutes']

model = LinearRegression()
model.fit(X, y)

# Printing the necessary values
coefficient = model.coef_[0]
print(f'Coefficient Estimate for Period: {coefficient}')

# Performing statistical tests on coefficients
# You can use hypothesis testing (t-test) on the coefficient to check significance
# Calculate the standard error, t-value, and p-value

# For the sake of example, let's assume you have standard errors (se) calculated
se = 0.1  # Replace with your calculated standard error

t_value = coefficient / se
p_value = 2 * (1 - stats.t.cdf(np.abs(t_value), len(combined_data) - 2))
print(f'P-value for Period: {p_value}')

# Interpret the results
if p_value < 0.05:  # Considering a significance level of 0.05
    print("The difference in viewing times between periods is statistically significant.")
else:
    print("There is no statistically significant difference in viewing times between periods.")
```
